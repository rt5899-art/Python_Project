# College Student Habits and Performance

This report analyzes the relationship between daily habits—including study time, sleep, technology usage, and physical activity—and the academic performance (GPA) of college students, based on typical patterns found in Kaggle datasets.

**Report: College Student Habits and Academic Performance1**

This analysis explores how daily habits and lifestyle choices impact the academic success (GPA) of university students. The findings indicate that study habits and physical activity have the strongest positive influence on GPA, while excessive screen time, social media usage, and late-night habits are significantly associated with lower academic performance. Contrary to some assumptions, part-time work does not strongly impact the median GPA, suggesting students manage to balance working hours.

**2. Key Findings & Correlations**

**A. Academic Performance vs. Study Habits** (Positive Correlation) Study Hours vs. GPA: There is a strong, direct positive correlation. Students who allocate more hours per day to studying consistently achieve a higher GPA.

**Attendance vs. GPA**: High attendance directly correlates with higher GPA, suggesting that active class participation is crucial for academic success. 

**Previous GPA**: Students with a higher previous GPA tend to maintain that trend, indicating the importance of established academic foundations. 

**Lifestyle Factors (Physical & Mental Health) Sleep Duration vs. GPA**: 

Students who sleep more (typically 7–9 hours) tend to have a higher GPA. There is a strong negative correlation between stress levels and sleep quality. 

**Physical Activity vs. GPA**: Higher levels of physical activity are positively associated with academic performance. 

**Stress Level:** High stress, often caused by poor time management or excessive workload, negatively impacts academic performance, leading to lower exam scores.

**Technology & Leisure Activities (Negative Correlation) Screen Time/Social Media vs. GPA**: High daily screen time, specifically non-academic usage (social media, Netflix, gaming), is negatively correlated with GPA. 

**Blue Square' Indicator**: In behavioral data, increased hours spent on social media, gaming, netflix_hours, screen_time, junk_food_frequency, and late_night_frequency are associated with lower academic performance. 

**Late Night Habits**: Frequent late-night studying or activities decrease sleep quality and are linked to lower GPAs. 

**Other Factors Part-time Hours vs. GPA**: The median GPA appears relatively consistent regardless of part-time work hours, suggesting students adapt their study schedules around employment.

```python
# Categorize based on procrastination and discipline
def categorize_habit(row):
    if row['procrastination_score'] > 7 and row['self_discipline'] < 4:
        return 'Last-Minute'
    elif row['weekly_study_sessions'] > 5 and row['self_discipline'] > 7:
        return 'Regular'
    else:
        return 'Average'

df['study_habit'] = df.apply(categorize_habit, axis=1)

# Visualize the impact on GPA
sns.boxplot(x='study_habit', y='gpa', data=df, palette='Set2', hue="study_habit")
plt.title('GPA by Study Habit (Regular vs. Last-Minute)')
plt.show()
```

# Conclusion & Recommendations Conclusion:

Academic performance is not merely a result of intelligence, but heavily influenced by daily habits. The optimal student profile for high GPA includes high attendance, regular study hours, adequate sleep, and high levels of physical activity. Recommendations for Students: Prioritize Sleep: Ensure at least 7 hours of sleep to manage stress and boost focus. Limit Screen Time: Reduce non-academic screen time to under 3 hours a day to prevent a significant drop in GPA. Optimize Study Time: Rather than cramming, consistent, daily study hours are more effective. Increase Physical Activity: Incorporate exercise to improve mental health and reduce anxiety. 

# Recommendations for Institutions: 

Implement Time Management Workshops: Address the high stress levels and poor sleep habits. Promote Digital Well-being: Create awareness campaigns about the negative impact of excessive non-academic screen time. Encourage Attendance: Incentivize attendance through active learning techniques.
